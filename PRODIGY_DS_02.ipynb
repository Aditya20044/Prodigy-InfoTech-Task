```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic_data = pd.read_csv('/Users/adityakumar/Downloads/titanic/train.csv')
```

```python
print(titanic_data.isnull().sum())
```

```python
titanic_data['Age'] = titanic_data['Age'].fillna(titanic_data['Age'].median())
```

```python
mode_embarked = titanic_data['Embarked'].mode()[0]
titanic_data['Embarked'] = titanic_data['Embarked'].fillna(mode_embarked)
```

```python
titanic_data.drop('Cabin', axis=1, inplace=True)
```

```python
print(titanic_data.isnull().sum())
```

```python
sns.countplot(x='Survived', data=titanic_data)
plt.title('Count of Survivors')
plt.show()
```

```python
sns.countplot(x='Survived', hue='Sex', data=titanic_data)
plt.title('Count of Survivors by Sex')
plt.show()
```

```python
sns.countplot(x='Survived', hue='Pclass', data=titanic_data)
plt.title('Count of Survivors by Passenger Class')
plt.show()
```

```python
sns.histplot(titanic_data['Age'], bins=20)
plt.title('Distribution of Age')
plt.show()
```

```python
sns.countplot(x='Embarked', data=titanic_data)
plt.title('Count of Passengers by Embarked Port')
plt.show()
```
