```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.preprocessing import LabelEncoder
```

```python
accident_data = pd.read_csv('/Users/adityakumar/Downloads/US_Accidents_March23.csv')
```

```python
print(accident_data.columns)
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(accident_data['Weather_Condition'])
plt.title('Distribution of Weather Conditions')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Wind_Speed(mph)', y='Severity', data=accident_data)
plt.title('Scatter Plot of Wind Speed vs. Accident Severity')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='Severity', data=accident_data)
plt.title('Distribution of Accident Severity')
plt.xlabel('Severity')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(accident_data['Visibility(mi)'], bins=20, kde=True)
plt.title('Distribution of Visibility')
plt.xlabel('Visibility (miles)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(accident_data['Temperature(F)'], bins=20, kde=True)
plt.title('Distribution of Temperature')
plt.xlabel('Temperature (Fahrenheit)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.scatterplot(x='Start_Lng', y='Start_Lat', data=accident_data, alpha=0.1)
plt.title('Accident Hotspots')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(y='Weather_Condition', data=accident_data, order=accident_data['Weather_Condition'].value_counts().iloc[:10].index)
plt.title('Top 10 Weather Conditions Contributing to Accidents')
plt.xlabel('Frequency')
plt.ylabel('Weather Condition')
plt.show()
```
